## Exhaustive search

```python
def exhaustive_search(function, lower, upper, precision):
    
    iterations = 0
    func_calls = 0
    
    n = int((upper-lower)/precision)
    x = [lower + k*(upper-lower)/n for k in range(n+1)]
    
    min_arg = 0
    min_func = 10000
    
    for xk in x:
        iterations += 1
        func_calls += 1
        if min_func > function(xk):
            min_arg = xk
            min_func = function(xk)
            
    print(f'Iterations: {iterations}')
    print(f'Function calls: {func_calls}')
            
    return min_arg, min_func
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *exhaustive_search(f1, lower, upper, precision)))
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *exhaustive_search(f2, lower, upper, precision)))
```

```python
lower = 0.1
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *exhaustive_search(f3, lower, upper, precision)))
```

## Dichotomy

```python
def dichotomy(function, lower, upper, precision):
    
    iterations = 0
    func_calls = 0
    sigma = precision/2

    while abs(lower - upper) > precision:

        x1 = (upper+lower-sigma)/2
        x2 = (upper+lower+sigma)/2


        if function(x1) >= function(x2):
            lower = x1
        else:
            upper = x2
        func_calls += 2
        iterations += 1

    extr = (lower + upper)/2
    
    print(f'Iterations: {iterations}')
    print(f'Function calls: {func_calls}')

    return extr, function(extr)
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *dichotomy(f1, lower, upper, precision)))
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *dichotomy(f2, lower, upper, precision)))
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *dichotomy(f3, lower, upper, precision)))
```

## Golden Section

```python
def golden_section(function, lower, upper, precision):
    
    iterations = 0
    func_calls = 0

    while abs(lower - upper) > precision:

        x1 = lower + (3-np.sqrt(5))/2*(upper-lower)
        x2 = upper + (-3+np.sqrt(5))/2*(upper-lower)


        if function(x1) >= function(x2):
            lower = x1
        else:
            upper = x2

        func_calls += 2
        iterations += 1

    extr = (lower + upper)/2
    
    print(f'Iterations: {iterations}')
    print(f'Function calls: {func_calls}')

    return extr, function(extr)
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *golden_section(f1, lower, upper, precision)))
```

```python
lower = 0
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *golden_section(f2, lower, upper, precision)))
```

```python
lower = 0.1
upper = 1
precision = 0.001

print("Min function in point {0:.4f}: {1:.4f}".format(
    *golden_section(f3, lower, upper, precision)))
```
